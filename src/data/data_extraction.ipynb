```python
import json


def extract_data(data_file):
    extracted_data = []
    with open(data_file, 'r') as f:
        data = json.load(f)

    for entry in data:
        try:
            soil_type = entry['input_fields']['soil_type']
            temperature = entry['input_fields']['temperature']
            geolocation = entry['input_fields']['geolocation']
            weather_conditions = entry['input_fields']['weather_conditions']
        except KeyError:
            
            print(f"Warning: Missing key(s) in entry {entry['combination']}")
            continue

        extracted_data.append({
            'soil_type': soil_type,
            'temperature': temperature,
            'geolocation': geolocation,
            'weather_conditions': weather_conditions
        })

    return extracted_data


extracted_data = extract_data('./soil.json')
print(f"Extracted data for {len(extracted_data)} entries:")


for i in extracted_data:
    print(i)
```
